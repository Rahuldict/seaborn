```python
# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
#"timepoint" and y = "signal" for different events and regions.
# Note: timepoint, signal, event, and region are columns in the fmri dataset.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Create a line plot using Seaborn
plt.figure(figsize=(10, 6))  # Set the figure size

# Use the sns.lineplot function to create the line plot
sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region")

# Add title and labels
plt.title("FMRI Signal Over Time")
plt.xlabel("Timepoint")
plt.ylabel("Signal")

# Show the legend
plt.legend(title="Event", loc="upper right")

# Show the plot
plt.show()
```

```python
 # Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
#'pclass', y = 'age' and y = 'fare'.
#Note: pclass, age, and fare are columns in the titanic dataset.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Create a figure with two box plots
plt.figure(figsize=(12, 5))

# Box plot for 'age' vs. 'pclass'
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic_data, x="pclass", y="age")
plt.title("Age vs. Pclass")
plt.xlabel("Pclass")
plt.ylabel("Age")

# Box plot for 'fare' vs. 'pclass'
plt.subplot(1, 2, 2)
sns.boxplot(data=titanic_data, x="pclass", y="fare")
plt.title("Fare vs. Pclass")
plt.xlabel("Pclass")
plt.ylabel("Fare")

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

```python
# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds_data = sns.load_dataset("diamonds")

# Create a histogram with hue for 'cut'
plt.figure(figsize=(10, 6))  # Set the figure size

# Use sns.histplot to create the histogram with hue
sns.histplot(data=diamonds_data, x="price", hue="cut", kde=True)

# Add title and labels
plt.title("Histogram of Diamond Prices by Cut")
plt.xlabel("Price")
plt.ylabel("Count")

# Show the legend
plt.legend(title="Cut")

# Show the plot
plt.show()
```

```python
 # Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "iris" dataset
iris_data = sns.load_dataset("iris")

# Create a pair plot with hue for 'species'
sns.set(style="ticks")  # Set the style for the pair plot
sns.pairplot(iris_data, hue="species")

# Show the plot
plt.show()
```
